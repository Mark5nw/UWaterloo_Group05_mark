Load wheatherData into a dataframe.
Make a copy of the raw dataframe to work on.

```python
import requests
import pandas as pd

url = 'https://api.weather.gc.ca/collections/climate-hourly/items?f=json&lang=en-CA&limit=50000&additionalProp1=%7B%7D&skipGeometry=false&offset=0&datetime=2020-01-01T00%3A00%3A00Z%2F2024-12-01T12%3A31%3A12Z&CLIMATE_IDENTIFIER=6144239'
r = requests.get(url)
x = r.json()
climateList = []
for record in x["features"]:
    climateList.append(record["properties"])
dfwheater_raw = pd.DataFrame(climateList)
print(dfwheater_raw)
```

Load collision data.
make a copy of the raw dataframe to work.

```python
base_url = "https://services1.arcgis.com/qAo1OsXi67t7XgmS/arcgis/rest/services/Traffic_Collisions/FeatureServer/0/query"
CollisionList = []
for year in range(2020, 2025):
    Collision_url = f"{base_url}?where=ACCIDENT_YEAR={year}&outFields=*&returnGeometry=false&outSR=4326&f=json"
    c = requests.get(Collision_url)
    y = c.json()
    for record in y["features"]:
        CollisionList.append(record["attributes"])

dfcollisions_raw = pd.DataFrame(CollisionList)
print(dfcollisions_raw)


```
